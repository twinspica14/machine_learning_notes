```python
XData = pd.DataFrame(y,columns=[1])
XData = XData.assign(k = np.arange(0,97))
XData['k'] = 1
XData.columns = [1,2]

XData = XData[[2,1]]
XData.columns = [1,2]
M = XData.as_matrix()
```

```python
Q = pd.DataFrame(z)
Q.columns = [1]
Q = Q.as_matrix()
```

```python
theta_init = np.zeros(shape=(2, 1))
iterations = 1500
alpha = 0.01 # Theta_init is matrix not data frame
# M[:,[0]] # how to access single column in matrix
```

```python
def gradientDescentLinear(M,Q,theta_init,alpha,iterations):
    X = M
    Y = Q
    lol = theta_init
    i = 0
    while(i<1500):
        err0 = (np.dot(X,lol)-Y)*X[:,[0]]
        err1 = (np.dot(X,lol)-Y)*X[:,[1]]
        lol[0][0] = lol[0][0] - alpha *(1/len(Y)) * np.sum(err0)
        lol[1][0] = lol[1][0] - alpha * (1/len(Y)) * np.sum(err1)
        i = i+1
        
    return lol
```

```python
kkk = gradientDescentLinear(M,Q,theta_init,alpha,iterations)
kkk
```
